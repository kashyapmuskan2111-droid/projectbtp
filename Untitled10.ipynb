```python
import pandas as pd
import zipfile

#extraction of zip file
with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    zip_ref.extractall('emotion_data')

#load dataframe
df = pd.read_csv('emotion_data/train.txt', sep=';', names=['Text', 'Emotion'])

print("Step 1 Complete: 16,000 sentences successfully loaded")
print(df.head())
```

```python
#preprocessing
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords')
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
  #remove noise and lowercase
    text = re.sub('[^a-zA-Z]', ' ', text).lower()
    words = text.split()
    # Stemming:
    return " ".join([ps.stem(w) for w in words if w not in stop_words])

df['Clean_Text'] = df['Text'].apply(clean_text)
print("Step 2 Complete: cleaning is done") #data normalization
```

```python

#Feature Extraction (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

tfidf = TfidfVectorizer(max_features=2000)
X = tfidf.fit_transform(df['Clean_Text']).toarray()

# BiLSTM takes 3D input: (Samples, Time-steps, Features)
X = np.reshape(X, (X.shape[0], 1, X.shape[1]))

# convert emotions into nos.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(df['Emotion'])

print("Step 3 Complete: Text successfully converted into nos. (Tensors) !")
```

```python
#BiLSTM Hybrid Model Training
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dense, Dropout

model = Sequential([
    Bidirectional(LSTM(64, activation='tanh'), input_shape=(1, 2000)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(len(le.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("Step 4: BiLSTM Training starts(Epochs: 10)...")
history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 1. Accuracy Graph
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('BiLSTM Model Accuracy')
plt.legend()
plt.show()

# 2. Confusion Matrix (Asli vs Predicted)
y_pred = np.argmax(model.predict(X), axis=1)
cm = confusion_matrix(y, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Result')
plt.show()
```
